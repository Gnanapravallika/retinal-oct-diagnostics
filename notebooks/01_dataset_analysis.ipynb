### Patient Leakage Verification
We verify that a patient's ID appears exclusively in one split (Train, Val, or Test).

```python
def check_patient_leakage(train_patients, val_patients, test_patients):
    intersection_1 = set(train_patients).intersection(set(val_patients))
    intersection_2 = set(train_patients).intersection(set(test_patients))
    intersection_3 = set(val_patients).intersection(set(test_patients))
    if len(intersection_1) == 0 and len(intersection_2) == 0 and len(intersection_3) == 0:
        print('✅ Success: Zero patient leakage detected across splits.')
    else:
        print('❌ Warning: Patient leakage detected!')

check_patient_leakage(['P01', 'P02'], ['P03'], ['P04'])
```
